```python
Gst.init(None)

VIDEO_FILE = Path("./PaddyThePelicanPiggyBankRobberyRestoredOriginal16MmPrint/Paddy the Pelican Piggy Bank Robbery [Restored Original 16 mm Print].mp4").resolve().as_uri()

# Simple GTK window
win = Gtk.Window()
win.connect("destroy", Gtk.main_quit)
win.set_default_size(800, 600)

video_area = Gtk.DrawingArea()
win.add(video_area)
win.show_all()

# Playbin with scaling
pipeline = Gst.parse_launch(
    f"playbin uri={VIDEO_FILE} video-sink='videoscale ! video/x-raw,width=800,height=600 ! ximagesink'"
)

pipeline.set_state(Gst.State.PLAYING)
Gtk.main()
```
